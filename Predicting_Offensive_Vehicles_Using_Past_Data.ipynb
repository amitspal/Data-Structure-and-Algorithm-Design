```python
class PoliceNode:
    def __init__(self,policeId,fineAmount):
        self.policeId = policeId
        self.fineAmount = fineAmount
        self.bookingCount = 1
        self.left = None
        self.right = None
                        
    def insertByPoliceId(self,policeId,fineAmount):
        if self.policeId is not None:
            if self.policeId > policeId:
                if self.left is None:
                    self.left = PoliceNode(policeId, fineAmount);
                else: 
                    self.left.insertByPoliceId(policeId, fineAmount)
            elif self.policeId < policeId:
                if self.right is None:
                    self.right = PoliceNode(policeId, fineAmount)
                else:
                    self.right.insertByPoliceId(policeId,fineAmount)
            else :
                self.bookingCount = self.bookingCount + 1
                self.fineAmount = self.fineAmount + fineAmount
        else :
            self.policeId = policeNode.policeId
            self.fineAmount = policeNode.fineAmount 
            
    def delete_node(self,policeId):
        if self is None:
            return(self)
        if self.policeId > policeId:
            self.left = delete_node(self.left,policeId)
        elif self.policeId < policeId:
            self.right = delete_node(self.right,policeId)
        else :
            if self.right is None:
                return self.left
            if self.left is None:
                return self.right
            
            temp_node = self.right
            while temp_node.left is not None:
                temp_node = tem_node.left
            self.right = delete_node(self.right,policeId)
            
        return(self)      
            
class ReorderedPoliceNode:
    def __init__(self,policeId,bookingCount,fineAmount):
        self.policeId = policeId
        self.fineAmount = fineAmount
        self.bookingCount = bookingCount
        self.left = None
        self.right = None
        
    def reorderPoliceTree(self,rootNode):
        if self.bookingCount >= rootNode.bookingCount:
            if self.left is None:
                self.left = ReorderedPoliceNode(rootNode.policeId,rootNode.bookingCount,rootNode.fineAmount);
            else: 
                self.left.reorderPoliceTree(rootNode)
            deletePoliceId = PoliceNode(self.left.policeId,self.left.fineAmount)
            deletePoliceId.delete_node(self.left.policeId)
        elif self.bookingCount < rootNode.bookingCount:
            if self.right is None:
                self.right = ReorderedPoliceNode(rootNode.policeId,rootNode.bookingCount,rootNode.fineAmount)
            else:
                self.right.reorderPoliceTree(rootNode)
            deletePoliceId = PoliceNode(self.right.policeId,self.right.fineAmount)
            deletePoliceId.delete_node(self.right.policeId)
                
def inorder(rootNode,output_Arr):
    if rootNode is not None:
        inorder(rootNode.left,output_Arr)
        output_Arr.append([rootNode.policeId,rootNode.bookingCount,rootNode.fineAmount])
        inorder(rootNode.right,output_Arr)   
        

def selectPoliceMan(rootNode,appendFile):   
    if rootNode is not None:
        selectPoliceMan(rootNode.left,appendFile)
        if(rootNode.bookingCount >= 10):
            appendFile.write(rootNode.policeId)
            appendFile.write("|")
            appendFile.write(str(rootNode.bookingCount))
            appendFile.write("\n")
        selectPoliceMan(rootNode.right,appendFile)

def printPoliceMan(rootNode,outputFilePath):
    appendFile = open(outputFilePath,'a')
    appendFile.write("------ Police List ------------ \n")
    appendFile.write("Police Id | Number of Bookings \n")
    selectPoliceMan(rootNode,appendFile)
    appendFile.close()

def sortTopTen(newRootNode,tempArr):
    if newRootNode is not None:
        sortTopTen(newRootNode.right,tempArr)
        if (len(tempArr) < 10):
            tempArr.append([newRootNode.policeId,newRootNode.bookingCount,newRootNode.fineAmount])
        sortTopTen(newRootNode.left,tempArr) 
        
def printTopTen(newRootNode,tempArr,outputFilePath):
    sortTopTen(newRootNode,tempArr)
    appendFile = open(outputFilePath,'a')
    appendFile.write("-------------- Police Top 10 ----------------\n")
    appendFile.write("PoliceId | no of bookings | total fine amount\n")
    for elements in tempArr:
        appendFile.write(elements[0])
        appendFile.write("|")
        appendFile.write(str(elements[1]))
        appendFile.write("|")
        appendFile.write(str(elements[2]))
        appendFile.write("\n")
      
    appendFile.close()  
        
def main():
    filePath = "C:/Users/amits/Desktop/BITS/DSAD_Assignment1_12-July-2020/"
    inputFilePath = filePath + "inputPS11.txt"
    outputFilePath = filePath + "police2.txt"
    print(inputFilePath)
    print(outputFilePath)
    readFile = open(inputFilePath, 'r')
    writeFile = open(outputFilePath,'w')

    policeId_fineAmount_Arr = []
    sorted_policeId_fineAmount_Arr = []
    sorted_bookingCount_Arr = []
    revised_policeId_fineAmount_Arr = []
    
    for fullLine in readFile.readlines():
        
        splitLine = fullLine.split("/")
        policeId = splitLine[0]
        fineAmount = float(splitLine[2].rstrip())
        
        policeId_fineAmount_Arr.append([policeId,fineAmount])
        
              
    print("Input Array :",policeId_fineAmount_Arr)                
   
    rootNode = PoliceNode(policeId_fineAmount_Arr[0][0],policeId_fineAmount_Arr[0][1])
         
    for i in range (1,len(policeId_fineAmount_Arr)):  
        rootNode.insertByPoliceId(policeId_fineAmount_Arr[i][0],policeId_fineAmount_Arr[i][1])
        
    inorder(rootNode,sorted_policeId_fineAmount_Arr)
        
    print("\n Sorted by police Id Array :",sorted_policeId_fineAmount_Arr) 
        
    newRootNode = ReorderedPoliceNode(sorted_policeId_fineAmount_Arr[0][0],sorted_policeId_fineAmount_Arr[0][1],sorted_policeId_fineAmount_Arr[0][2])
           
    for i in range (1,len(sorted_policeId_fineAmount_Arr)): 
        nextNode = ReorderedPoliceNode(sorted_policeId_fineAmount_Arr[i][0],sorted_policeId_fineAmount_Arr[i][1],sorted_policeId_fineAmount_Arr[i][2])
        newRootNode.reorderPoliceTree(nextNode)
        
    inorder(newRootNode,sorted_bookingCount_Arr) 
    
    print("\n Sorted by booking count Array :",sorted_bookingCount_Arr) 
    
    printPoliceMan(newRootNode,outputFilePath)
    
    tempArr = []
    printTopTen(newRootNode,tempArr,outputFilePath)
    
    readFile.close()
    writeFile.close()
        
    
```
